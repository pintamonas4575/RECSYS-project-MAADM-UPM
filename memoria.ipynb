```python
def similarity(a: np.ndarray, b: np.ndarray) -> float: #OBLIGATORIO NUMPY
    a_index = np.where(a!=0)
    b_index = np.where(b!=0)
    intersect = np.intersect1d(a_index, b_index)
    if len(intersect)==0:
        return 0
    union = np.union1d(a_index, b_index)
    factor = len(intersect) / len(union)
    msd = 1/len(intersect) * np.sum((a[intersect]-b[intersect])**2)
    return factor*msd  
```
