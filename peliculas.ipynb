## DATOS

```python
movies = pd.read_csv('data/ml-100k/u.item', sep='|', encoding='latin-1', header=None)
movies.head()
```

```python
ratings = pd.read_csv("data/ml-100k/u.data", sep="\t", names=["user_id", "movie_id", "rating", "timestamp"])
ratings.info()
ratings.head()
```

```python
from scipy.sparse import csr_matrix

# Obtener dimensiones de la matriz usando count
num_users = ratings["user_id"].nunique()
num_movies = ratings["movie_id"].nunique()

print(f"Number of users: {num_users}")
print(f"Number of movies: {num_movies}")

# Crear la matriz dispersa en formato CSR (Compressed Sparse Row)
sparse_matrix = csr_matrix(
    (ratings["rating"], (ratings["user_id"] - 1, ratings["movie_id"] - 1)), 
    shape=(num_users, num_movies)
)

# Mostrar información de la matriz dispersa
print("Shape: " + str(sparse_matrix.shape))  # Dimensiones (número de usuarios, número de películas)
print("Total de casillas: "+ str(sparse_matrix.shape[0] * sparse_matrix.shape[1]))  # Número total de casillas
print("Not empty values: " + str(sparse_matrix.nnz))    # Número de elementos no nulos
```

```python
user_id = 1
user_ratings = sparse_matrix[user_id - 1, :]  # Restamos 1 porque los índices en la matriz comienzan desde 0
# Obtener los índices de las películas calificadas por el usuario
rated_movie_indices = user_ratings.indices

# Mostrar los títulos de las películas calificadas
rated_movie_titles = movies.iloc[rated_movie_indices, 1]

print(f"User {user_id} has rated the following movies:")
for title, rating in zip(rated_movie_titles, user_ratings.data):
    print(f"{title}: {rating}")
```
